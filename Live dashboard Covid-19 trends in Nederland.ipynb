Deel 1: de datasets prepareren (inlezen en op jaarbasis splitsen).

```python
# Pandas importeren
import pandas as pd

# de dataset van OU inlezen
dataset_OU = pd.read_csv('dataset/COVID-19_aantallen_gemeente_per_dag.csv', header=0, sep=';')
dataset_OU.head()
```

```python
# de dataset van OU checken: 
# - Date_of_publication, Province, Total_reported, Hospital_admission en Deceased hebben geen missing waarden,
# - Municipality_code en Municipality_name hebben wel missing waarden.
dataset_OU.info()
```

```python
# de dataset van OU op jaarbasis opsplitsen zodat voor elk jaar een afzonderlijke dataset wordt gecreeerd.
dataset_OU['year'] = dataset_OU['Date_of_publication'].str[:4]

# dataset voor 2020 aanmaken
dataset_OU_2020 = dataset_OU[dataset_OU['year']=='2020']
dataset_OU_2020.head()
```

```python
# dataset voor 2021 aanmaken
dataset_OU_2021 = dataset_OU[dataset_OU['year']=='2021']
dataset_OU_2021.head()
```

Deel 2: de datasets bewerken tbv data visualisatie.

```python
# selecteer de relevante informatie subset
temp_dataset = dataset_OU_2020[['Hospital_admission','Deceased','Total_reported','Province']]

# tellen de aantallen op per Hospital_admission, Deceased en Total_reported, gegroepeerd van Province
temp_dataset = temp_dataset.groupby(['Province'])[['Hospital_admission','Deceased','Total_reported']].sum()

# zet de tijdelijke subset om naar province_2020 en past de pandas melt functie toe, ter voorbereiding van de visualisatie
province_2020 = temp_dataset.reset_index()
province_2020 = pd.melt(province_2020, id_vars='Province', var_name='Type', value_name='Numbers')
province_2020['Year']='2020' # toegevoegd om de landelijke telling mogelijk te maken
province_2020.head(15)
```

```python
# voer dezelfde stappen uit voor de dataset van 2021

# selecteer de relevante informatie subset
temp_dataset = dataset_OU_2021[['Hospital_admission','Deceased','Total_reported','Province']]

# tellen de aantallen op per Hospital_admission, Deceased en Total_reported, gegroepeerd van Province
temp_dataset = temp_dataset.groupby(['Province'])[['Hospital_admission','Deceased','Total_reported']].sum()

# zet de tijdelijke subset om naar province_2020 en past de pandas melt functie toe, ter voorbereiding van de visualisatie
province_2021 = temp_dataset.reset_index()
province_2021 = pd.melt(province_2021, id_vars='Province', var_name='Type', value_name='Numbers')
province_2021['Year']='2021' # toegevoegd om de landelijke telling mogelijk te maken
province_2021.head(15)
```

```python
# creer een dictionary zodat de juiste dataset obv het ingevoerde jaar dynamisch kan worden geselecteerd
datasets = {}
datasets['2020'] = province_2020
datasets['2021'] = province_2021
```

Deel 3: functies aanmaken tbv data visualisatie.

```python
# importeer numpy, seaborn en matplotlib tbv datavisualisatie
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# de functie om de eerste screenshot van opdracht 2 te tonen
def show_national_graph(year='2020', hue_order=['Total_reported','Hospital_admission','Deceased']):
    
    # toon de barplot conform de eerste screenshot van opdracht 2
    g = sns.barplot(x = 'Year',
            y = 'Numbers',
            hue = 'Type',
            hue_order=hue_order,
            estimator = np.sum,
            ci=None,
            data = datasets[year])
    
    # de screenshots van opdracht 2 bevatten geen xlabel en ylabel, daarom deze hier ook verwijderen.
    g.set(xlabel='', ylabel='')

    # de eerste screenshot van opdracht 2 heeft een titel, daarom deze hier ook toegevoegd.
    g.set(title='The Netherlands in ' + year)

    # de screenshots van opdracht 2 bevatten ook geen titel voor de legenda, daarom deze hier ook verwijderen.
    plt.legend(title='')

    # de province namen geroteerd voor de leesbaarheid, conform de screenshots van opdracht 2
    plt.xticks(rotation=45)

    plt.show()
```

```python
# de functie om de tweede screenshot van opdracht 2 te tonen
def show_province_graph(year='2020', hue_order=['Total_reported','Hospital_admission','Deceased']):
    
    # toon de barplot conform de tweede screenshot van opdracht 2
    g = sns.barplot(x = 'Province',
                y = 'Numbers',
                hue = 'Type',
                hue_order=hue_order,
                data = datasets[year])

    # de screenshots van opdracht 2 bevatten geen xlabel en ylabel, daarom deze hier ook verwijderen.
    g.set(xlabel='', ylabel='')

    # de screenshots van opdracht 2 bevatten ook geen titel voor de legenda, daarom deze hier ook verwijderen.
    plt.legend(title='')

    # de province namen geroteerd voor de leesbaarheid, conform de screenshots van opdracht 2
    plt.xticks(rotation=45)

    plt.show()
```

Deel 4: GUI componenten en de event_handler functie aanmaken om obv de gebruikersinvoer de juiste visualisatie te tonen

```python
import ipywidgets as widgets
```

```python
# deze functie is de brug tussen de gebruikersinvoer en de visuatie die getoond moet worden
def event_handler(year='2020', total_reported=True, hospital_admission=True, deceased=True, grouping='Netherlands'):
    # Afhankelijk van de gebruikersinvoer, de juiste datasoort (Total_reported, Hospital_admission en Deceased) tonen
    hue_order=[]
    if total_reported:
        hue_order.append('Total_reported')
    if hospital_admission:
        hue_order.append('Hospital_admission')
    if deceased:
        hue_order.append('Deceased')

    # als geen opties zijn geselecteerd, dan alle opties tonen, om Python exception te voorkomen
    if len(hue_order) == 0:
        hue_order=['Total_reported','Hospital_admission','Deceased']
        
    # Afhankelijk van de gebruikersinvoer, de juiste groepering (landelijk of per province) tonen    
    if grouping =='Netherlands':
        show_national_graph(year,hue_order)
            
    if grouping =='All provinces':
        show_province_graph(year,hue_order)
            
```

```python
# deze functie toont de GUI componenten en geeft de gebruikersinvoer door aan de event_handler

# dropdown opties om het te tonen jaar te selecteren
year_selector = widgets.Dropdown(
    options=datasets.keys(), # alleen de opties tonen waar datasets beschikbaar voor zijn
    value='2020',
    description='year',
    disabled=False
    )

# checkbox tonen voor Total_reported
total_reported_selector = widgets.Checkbox(
    value=True,
    description='Total_reported',
    disabled=False,
    indent=True
)

# checkbox tonen voor Hospital_admission
hospital_admission_selector = widgets.Checkbox(
    value=True,
    description='Hospital_admission',
    disabled=False,
    indent=True
)

# checkbox tonen voor Deceased
deceased_selector = widgets.Checkbox(
    value=True,
    description='Deceased',
    disabled=False,
    indent=True
)

# dropdown opties om het te tonen groepering te selecteren
grouping_selector = widgets.Dropdown(
    options=['Netherlands','All provinces'],
    value='Netherlands',
    description='province',
    disabled=False
    )

# interactieve dataselectie en datavisualisatie
widgets.interact(event_handler, year=year_selector, total_reported=total_reported_selector, hospital_admission=hospital_admission_selector, deceased=deceased_selector, grouping=grouping_selector)
```

```python
# Groepeer de data per jaar en maand, ter voorbereiding van de time series visualisatie
dataset_OU = dataset_OU[['Date_of_publication','Province','Total_reported','Hospital_admission','Deceased']]
dataset_OU['Timing'] = dataset_OU['Date_of_publication'].str[:7]
dataset_OU.head()
```

```python
# Maak de dataset aan ter voorbereiding van de timeseries visualisatie
provinces = list(dataset_OU['Province'].unique())  # maak de lijst van provinces aan die in de dataset beschikbaar zijn

# creer een dictionary zodat de juiste dataset obv het ingevoerde province dynamisch kan worden geselecteerd
datasets_TimeSeries = {}  

for province in provinces:
    # selecteer de relevante informatie subset
    temp_dataset = dataset_OU[dataset_OU['Province']==province]
    temp_dataset = temp_dataset[['Total_reported','Hospital_admission','Deceased','Timing']]
    
    # pas de pandas melt functie toe, ter voorbereiding van de visualisatie
    temp_dataset_TimeSeries = pd.melt(temp_dataset, id_vars='Timing', var_name='Type', value_name='Numbers')
    
    # neem de dataset in de dictionary op, gekoppeld aan de province
    datasets_TimeSeries[province]=temp_dataset_TimeSeries
```

```python
# de functie om de timeseries visualisatie te tonen
def show_timeSeries_graph(province='Groningen', hue_order=['Total_reported','Hospital_admission','Deceased']):

    # toon de timeseries visualisatie 
    g = sns.relplot(x='Timing', 
                    y='Numbers', 
                    data=datasets_TimeSeries[province], 
                    kind='line', 
                    estimator=np.sum,hue='Type', 
                    hue_order=hue_order,
                    ci=None)

    # de screenshots van opdracht 2 bevatten geen xlabel en ylabel, daarom deze hier ook verwijderen.
    g.set(xlabel='', ylabel='')

    # de screenshots van opdracht 2 bevatten ook geen titel voor de legenda, daarom deze hier ook verwijderen.
    plt.legend(title='')
    g._legend.remove() # voorkomen dat de tweede legenda wordt getoond.
    
    # timing geroteerd voor de leesbaarheid
    plt.xticks(rotation=90)
    
    plt.show()
```

```python
# deze functie is de brug tussen de gebruikersinvoer en de visuatie die getoond moet worden, om de timeseries data te tonen
def event_handler_TimeSeries(province='Groningen', total_reported=True, hospital_admission=True, deceased=True):
    # Afhankelijk van de gebruikersinvoer, de juiste datasoort (Total_reported, Hospital_admission en Deceased) tonen
    hue_order=[]
    if total_reported:
        hue_order.append('Total_reported')
    if hospital_admission:
        hue_order.append('Hospital_admission')
    if deceased:
        hue_order.append('Deceased')

    # als geen opties zijn geselecteerd, dan alle opties tonen, om Python exception te voorkomen
    if len(hue_order) == 0:
        hue_order=['Total_reported','Hospital_admission','Deceased']
        
    # Afhankelijk van de gebruikersinvoer, de timeseries data van de geselecteerde province te tonen
    show_timeSeries_graph(province, hue_order)
```

```python
# deze functie toont de GUI componenten en geeft de gebruikersinvoer door aan de event_handler om de timeseries data te tonen

# dropdown opties om het te tonen province te selecteren
province_selector = widgets.Dropdown(
    options=datasets_TimeSeries.keys(), # alleen de opties tonen waar datasets beschikbaar voor zijn
    value='Groningen',
    description='province',
    disabled=False
    )

# checkbox tonen voor Total_reported
total_reported_selector = widgets.Checkbox(
    value=True,
    description='Total_reported',
    disabled=False,
    indent=True
)

# checkbox tonen voor Hospital_admission
hospital_admission_selector = widgets.Checkbox(
    value=True,
    description='Hospital_admission',
    disabled=False,
    indent=True
)

# checkbox tonen voor Deceased
deceased_selector = widgets.Checkbox(
    value=True,
    description='Deceased',
    disabled=False,
    indent=True
)

# interactieve dataselectie en datavisualisatie
widgets.interact(event_handler_TimeSeries, province=province_selector, total_reported=total_reported_selector, hospital_admission=hospital_admission_selector, deceased=deceased_selector)
```
